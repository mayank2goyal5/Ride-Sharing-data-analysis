```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("ride_sharing_raw_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['rating']=df['rating'].fillna(df['rating'].median())
```

```python
df.isnull().sum()
```

```python
df=df[df['fare_amount']>0]
```

```python
df=df[df['distance_km']>0]
```

```python
df['pickup_datetime']=pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime']=pd.to_datetime(df['dropoff_datetime'])
df['trip_duration']=(df['dropoff_datetime'] - df['pickup_datetime']).dt.total_seconds()/60
```

```python
df=df[df['trip_duration']<180]
```

```python
df = df[~((df['pickup_location'] != df['dropoff_location']) & (df['distance_km'] < 50))]
```

```python
df
```

```python
df['pickup_location']=df['pickup_location'].str.title()
df['dropoff_location']=df['dropoff_location'].str.title()
```

```python
df['hour']=df['pickup_datetime'].dt.hour
df['day']=df['pickup_datetime'].dt.day_name()
df['month']=df['pickup_datetime'].dt.month

df['is_weekend']=df['day'].isin(['Saturday','Sunday']).astype(int)
df['revenue']=df['fare_amount']*df['surge_multiplier']
df['revenue_per_km']=df['revenue']/df['distance_km']
```

```python
df.info()
df.describe()
```

```python
df.head()
```

```python
df.to_csv("cleaned_rides.csv", index=False)
```
